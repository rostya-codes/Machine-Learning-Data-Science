```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(4, 10), dpi=200)

axes[0][0].plot(x, y)
axes[0][0].set_ylabel('Y label')
axes[0][1].plot(a, b)
axes[1][0].plot(x*2, y*1.5)
axes[1][0].set_title('TITLE 1.1')
axes[1][1].plot(a*2, b*1.5)

fig.subplots_adjust(wspace=1, hspace=1)  # Расстояния между графиками в %. *Перекрывается plt.tight_layout()
fig.suptitle('Figure Level Title')  # Super title
plt.tight_layout()  # Поровнять
fig.set_figwidth(10)
fig.savefig('new_subplots.png', bbox_inches='tight')
```
