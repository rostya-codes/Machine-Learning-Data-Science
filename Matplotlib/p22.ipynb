```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
x
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

# ax.plot(x, x, color='#f4a839', lw=10, ls='-.')  # lw - linewidth

lines = ax.plot(x, x+1, color='purple', linewidth=2, linestyle='--', marker='o', ms=10)
lines[0].set_dashes([5, 2, 10, 2, 20, 6])  # Стилизация пунктира

lines = ax.plot(x, x+2, color='orange', linewidth=2, linestyle='--', marker='v', ms=5,
               markerfacecolor='orange', markeredgewidth=8, markeredgecolor='red')

ax.legend()
```
