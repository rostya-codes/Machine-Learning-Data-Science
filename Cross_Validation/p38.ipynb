```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../UNZIP_ME_FOR_NOTEBOOKS_ML_RUS_V7/DATA/Advertising.csv')
```

```python
df.head()
```

```python
X = df.drop('sales', axis=1)
```

```python
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_other, y_train, y_other = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# test_size=0.5 (50% из 30% other ---> test_size=15% от исходного объёма данных)

X_eval, X_test, y_eval, y_test = train_test_split(X_other, y_other, test_size=0.5, random_state=101)
```

```python
len(df)  # 100% --- global
```

```python
len(X_train)  # 70% --- train
```

```python
len(X_eval)  # 15% --- validation
```

```python
len(X_test)  # 15% --- test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_eval = scaler.transform(X_eval)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model_1 = Ridge(alpha=100)
```

```python
model_1.fit(X_train, y_train)
```

```python
y_eval_pred = model_1.predict(X_eval)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_eval, y_eval_pred)
```

```python
model_2 = Ridge(alpha=1)
```

```python
model_2.fit(X_train, y_train)
```

```python
new_pred_eval = model_2.predict(X_eval)
```

```python
mean_squared_error(y_eval, new_pred_eval)
```

```python
y_final_test_pred = model_2.predict(X_test)
```

```python
mean_squared_error(y_test, y_final_test_pred)
```
