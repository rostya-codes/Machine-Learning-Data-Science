```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../UNZIP_ME_FOR_NOTEBOOKS_ML_RUS_V7/DATA/Advertising.csv')
```

```python
df.head()
```

```python
## Создаём X и y
X = df.drop('sales',axis=1)
y = df['sales']

# Разбиение на TRAIN и TEST
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Масштабирование данных (SCALE)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)  # Hold out test set
```

```python
from sklearn.linear_model import ElasticNet
```

```python
help(ElasticNet)
```

```python
base_elastic_net_model = ElasticNet()
```

```python
param_grid = {'alpha': [0.1, 1, 5, 50, 100],
              'l1_ratio': [.1, .5, .7, .95, .99, 1]}
# Гиперпараметры. Ключи должны быть указаны такие как в параметрах модели.
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_net_model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5, verbose=2)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_params_  # Наилучшая комбинация гиперпараметров
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)  # Final
```
