```python
## Создаём X и y
X = df.drop('sales',axis=1)
y = df['sales']

# Разбиение на TRAIN и TEST
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Масштабирование данных (SCALE)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)  # Hold out test set
```

```python
from sklearn.model_selection import cross_validate  # С помощью этого медота можно получить больше информации
```

```python
model = Ridge(alpha=100)
```

```python
scores = cross_validate(model, X_train, y_train,
                        scoring=['neg_mean_squared_error', 'neg_mean_absolute_error'],
                        cv=10)
```

```python
scores
```

```python
scores = pd.DataFrame(scores)
```

```python
scores
```

```python
scores.mean()
```

```python
model = Ridge(alpha=1)
```

```python
scores = cross_validate(model, X_train, y_train,
                        scoring=['neg_mean_squared_error', 'neg_mean_absolute_error'],
                        cv=10)
```

```python
scores = pd.DataFrame(scores)
```

```python
scores.mean()
```

```python
model.fit(X_train, y_train)
```

```python
y_final_pred = model.predict(X_test)
```

```python
mean_squared_error(y_test, y_final_pred)  # Работа модели на абсолютно новых данных
```
