```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
#help(pd.merge)
```

```python
pd.merge(registrations, logins, how='inner', on='name')  # Слияние, порядок таблиц неважен
```

```python
pd.merge(logins, registrations, how='inner', on='name')  # Входящие id поменялись местами
```

```python
pd.merge(left=registrations, right=logins, how='left', on='name')  # Порядок параметров left and right ВАЖЕН
```

```python
pd.merge(left=registrations, right=logins, how='right', on='name')  # Данные из how= в любом случае должны попасть в результат
```

```python
pd.merge(left=registrations, right=logins, how='right')  # Не указывать параметр on= нежелательно
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations, logins, how='outer', on='name')  # Порядок указаных таблиц неважен. Получаем имена из обеих таблиц, при этом в пустые строки попадает значение NaN
```

```python
registrations = registrations.set_index('name')
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations, logins, left_index=True, right_on='name', how='inner')
```

```python
registrations = registrations.reset_index()  # Сброс индекса
registrations
```

```python
registrations.columns = ['reg_name', 'reg_id']
```

```python
registrations
```

```python
results = pd.merge(registrations, logins, how='inner', left_on='reg_name', right_on='name')
```

```python
results
```
