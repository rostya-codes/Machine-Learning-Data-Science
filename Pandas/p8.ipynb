```python
import timeit
```

```python
# этот фрагмент кода будет запускаться только один раз
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('/home/rostya/AI_course/UNZIP_ME_FOR_NOTEBOOKS_ML_RUS_V7/03-Pandas/tips.csv')
def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
'''
```

```python
# это фрагменты кода, для которых будет измеряться время выполнения
# (они будут выполняться много раз)
stmt_one = ''' # Добавление новой колонки через lambda
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
'''

stmt_two = ''' # Добавление новой колонки через np.vectorize
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
```

```python
timeit.timeit(setup=setup, stmt=stmt_one,number=10000)
```

```python
timeit.timeit(setup=setup, stmt=stmt_two,number=10000)
```
