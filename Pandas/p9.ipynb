```python
help(df.dropna)
```

```python
df.dropna()  # Удалит все строки в которых есть хотя бы одно неопределённое значение
```

```python
df.dropna(thresh=1)  # В threshhold передается число минимального количества заполненных данных в столбце от которого будет производиться вывод
```

```python
df
```

```python
df.dropna(axis=1)  # Удалить все колонки которые содержат NaN, а в датафрейме в каждой колонке есть хотя бы одно такое значение
```

```python
df.dropna(subset=['last_name'])  # Удалит только те строки в которых значение NaN встречается в колонке last_name
```

```python
df
```

```python
help(df.fillna)
```

```python
df.fillna('NEW_VALUE')  # Подставить это значение вместо NaN
```

```python
df
```

```python
df['pre_movie_score'] = df['pre_movie_score'].fillna(0)  # Подставить 0 вместо NaN
df
```

```python
df['pre_movie_score'].mean()  # Среднее значение на основе только значемых значений
```

```python
df['pre_movie_score'].fillna(df['pre_movie_score'].mean())
```

```python
df.fillna(df.mean(numeric_only=True))  # Замена числовых значений на средние значения
```
