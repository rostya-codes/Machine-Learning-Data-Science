```python
def percent_missing(data):
    result = 100 * data.isnull().sum() / len(df)
    result = result[result > 0].sort_values()
    return result
```

```python
percent_nan = percent_missing(df)
```

```python
percent_nan
```

```python
plt.figure(figsize=(7, 3), dpi=200)
sns.barplot(x=percent_nan.index, y=percent_nan)
plt.xticks(rotation=90)
plt.ylim(0, 1)  # Значения от 0 до 1
```

```python
# для числовых колонок - fillna 0
bsmt_num_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF','Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[bsmt_num_cols] = df[bsmt_num_cols].fillna(0)  # Замена отсутствующих числовых значений на 0

# для текстовых колонок - fillna None
bsmt_str_cols =  ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
df[bsmt_str_cols] = df[bsmt_str_cols].fillna('None')  # Замена отсутствующих текстовых значений на None
```
