```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/rostya/projects/python/github/ai/Machine-Learning-Data-Science/UNZIP_ME_FOR_NOTEBOOKS_ML_RUS_V7/08-Linear-Regression-Models/Advertising.csv')
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
df.head()
```

```python
sns.regplot(data=df, x='total_spend', y='sales')  # Чем больше денег тратится на рекламу тем больше продажи
```

```python
x = df['total_spend']
y = df['sales']
```

```python
# y = mx+b
# y = B1*x + B0
help(np.polyfit)
```

```python
np.polyfit(x, y, deg=1)
```

```python
sns.scatterplot(data=df, x='total_spend', y='sales')
plt.plot(potential_spend, predicted_sales, color='red')
```
