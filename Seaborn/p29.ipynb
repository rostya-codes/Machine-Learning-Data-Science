```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/rostya/projects/python/github/ai/Machine-Learning-Data-Science/UNZIP_ME_FOR_NOTEBOOKS_ML_RUS_V7/05-Seaborn/country_table.csv')
```

```python
df = df.set_index('Countries')
```

```python
# Heatmap
plt.figure(dpi=200)
sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True,
           cmap='viridis', center=30)  # cmap = palette
```

```python
# Clustermap
plt.figure(dpi=200)
sns.clustermap(data=df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True,
           cmap='viridis', center=30, col_cluster=False)  # cmap = palette
```
