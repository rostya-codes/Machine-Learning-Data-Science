```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('/home/rostya/projects/python/github/ai/Machine-Learning-Data-Science/UNZIP_ME_FOR_NOTEBOOKS_ML_RUS_V7/05-Seaborn/dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(20, 5), dpi=200)
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
sns.set(style='whitegrid')
sns.displot(data=df, x='salary', bins=20, color='red', edgecolor='orange', linewidth=2)
```

```python
sns.histplot(data=df, x='salary')
```

```python
sns.kdeplot(data=df, x='salary')
```
