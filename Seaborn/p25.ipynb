```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/rostya/projects/python/github/ai/Machine-Learning-Data-Science/UNZIP_ME_FOR_NOTEBOOKS_ML_RUS_V7/05-Seaborn/dm_office_sales.csv')
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df, x='level of education', hue='division', palette='Set2')  # hue= - разбивка на подкатегории
```
