```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/rostya/projects/python/github/ai/Machine-Learning-Data-Science/UNZIP_ME_FOR_NOTEBOOKS_ML_RUS_V7/05-Seaborn/StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='kde', shade=True)
# kind='hex' - шестиуголизация, kind='kde' - KDE style
```

```python
sns.jointplot(data=df, x='math score', y='reading score', hue='gender')
plt.legend(loc=(1.25, 0.15))
```

```python
# Pairplot
sns.pairplot(data=df, hue='gender', corner=True)  # corner=True - оставить уникальные графики
```
