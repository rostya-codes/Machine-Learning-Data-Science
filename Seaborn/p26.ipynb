```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/rostya/projects/python/github/ai/Machine-Learning-Data-Science/UNZIP_ME_FOR_NOTEBOOKS_ML_RUS_V7/05-Seaborn/StudentsPerformance.csv')
```

```python
df.head()
```

```python
# Boxplot
plt.figure(figsize=(10,8), dpi=200)
sns.boxplot(data=df, x='reading score', y='parental level of education', hue='test preparation course', palette='Set2')
plt.legend(loc=(1.05, 0.5))
```

```python
# Violinplot
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education',
               palette='Set2', split=True,
               inner='stick', bw=0.1)  # bw= - чувствительность
plt.legend(loc=(1.05, 0.5))
```

```python
# Swarmplot
plt.figure(figsize=(10, 5), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course',
             dodge=True)
plt.legend(loc=(1.05, 0.5))
```

```python
# Boxenplot
sns.boxenplot(data=df, x='math score', y='test preparation course', hue='gender')
```
